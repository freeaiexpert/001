```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset from Seaborn
titanic_data = sns.load_dataset('titanic')

# ---------------------------- Part 1: Boxplot for Age by Gender and Survival ----------------------------
# Create a boxplot for the distribution of 'age' by 'sex' and 'survived'
plt.figure(figsize=(10, 6))  # Set the figure size
sns.boxplot(x='sex', y='age', hue='survived', data=titanic_data, palette='coolwarm', showfliers=False)  # Boxplot
plt.title('Distribution of Age by Gender and Survival Status', fontsize=16)  # Set the title
plt.xlabel('Gender', fontsize=12)  # Set the x-axis label
plt.ylabel('Age', fontsize=12)  # Set the y-axis label
plt.show()  # Display the plot

# ---------------------------- Observations ----------------------------

# Observations and Inference:
# 1. **Male Passengers**:
#    - The age distribution for male passengers has a wider spread, with a median around 30 years.
#    - There is a noticeable difference between male passengers who survived and those who did not.
#    - The passengers who survived tend to be younger, as seen by the lower median age for the 'survived' group.

# 2. **Female Passengers**:
#    - Female passengers tend to have a narrower age distribution, with a median closer to 25 years.
#    - Similar to male passengers, female passengers who survived tend to be younger on average.
#    - The difference between the age distribution of female survivors and non-survivors is less pronounced than for male passengers.

# 3. **Effect of Survival on Age**:
#    - For both genders, younger passengers had a higher chance of surviving.
#    - This trend is more prominent among female passengers, where the survivors are younger on average.
#    - The non-survivors, especially among males, have a wider spread across ages.

# Key Insights:
# - **Gender and Survival**: Female passengers had a higher chance of survival than male passengers, consistent with the historical record.
# - **Age and Survival**: Younger passengers were more likely to survive. This trend was more pronounced in female passengers.
# - **Survival Rates by Gender**: The plot indicates that women had a significantly higher survival rate than men during the Titanic disaster.
```
