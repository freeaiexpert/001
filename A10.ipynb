```python
# Import necessary libraries
import pandas as pd  # For data manipulation
import seaborn as sns  # For statistical data visualization
import matplotlib.pyplot as plt  # For plotting graphs

# Load the Iris dataset from Seaborn
iris_data = sns.load_dataset('iris')

# Display the first few rows of the dataset to understand its structure
print(iris_data.head())
# This gives an overview of the dataset, showing the first 5 rows. It will display the features: sepal_length, sepal_width, petal_length, petal_width, and species.

# Show basic information about the dataset like data types and non-null counts
print(iris_data.info())
# This will display the column names along with their data types and the number of non-null values in each column. It's important to check if the dataset has any missing values.

# List the features and their types
features_types = iris_data.dtypes
print("\nFeatures and their types:")
print(features_types)
# This will show the data types of each feature. Typically, features like 'sepal_length', 'sepal_width', 'petal_length', and 'petal_width' are continuous (numeric), while 'species' is a categorical feature (nominal).

# ---------------------------- Part 1: Histogram of Each Feature ----------------------------

# Create histograms for each feature to illustrate feature distributions
iris_data.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle('Histograms of Iris Dataset Features', fontsize=16)  # Set the title for all histograms
plt.tight_layout()  # Automatically adjusts layout to avoid overlapping
plt.show()
# Histograms for each feature ('sepal_length', 'sepal_width', 'petal_length', 'petal_width') will show how data is distributed across each feature.
# The 'bins' argument controls the number of bins for the histogram, and 'edgecolor' adds a border to each bin for clarity.

# ---------------------------- Part 2: Boxplot for Each Feature ----------------------------

# Create boxplots for each feature in the dataset to check for outliers and distribution
plt.figure(figsize=(12, 8))  # Set the figure size
sns.boxplot(data=iris_data, orient='h', palette='Set2')  # Create horizontal boxplots for each feature
plt.title('Boxplot for Each Feature in Iris Dataset', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.show()
# A boxplot for each feature provides information about the median, interquartile range (IQR), and potential outliers in the data.
# The 'orient' argument makes the boxplots horizontal, and 'palette' defines the colors used.

# ---------------------------- Inferences ----------------------------

# Observations:
# 1. **Features and Types**:
#    - 'sepal_length', 'sepal_width', 'petal_length', 'petal_width' are continuous numeric features (floats).
#    - 'species' is a categorical variable representing the flower's species (setosa, versicolor, virginica).
#
# 2. **Histogram Observations**:
#    - 'sepal_length' and 'petal_length' seem to have a near-normal distribution, with most values clustered around the mean.
#    - 'sepal_width' and 'petal_width' are more skewed and have distinct distribution shapes for each species.
#    - The histograms show that the dataset has a good representation of each feature across different species.
#
# 3. **Boxplot Observations**:
#    - **Outliers**: The boxplots help identify any outliers in the data. For example, in the 'sepal_width' boxplot, there might be a potential outlier in the 'setosa' species.
#    - The distributions of the features differ between species, for example, the 'petal_length' feature is much larger for 'versicolor' and 'virginica' compared to 'setosa'.
#
# Key Insights:
# - **Sepal and Petal Length**: The 'petal_length' and 'sepal_length' features are the primary discriminators between species, as their distributions differ substantially between 'setosa', 'versicolor', and 'virginica'.
# - **Outliers**: T
```
